# Comparative Figures based on the dataset

### Figure showing the Number of wins per driver:

```python
Driver_Wins_fig = px.bar(F1Drivers_df, x='Driver', y='Race_Wins', 
             title='Number of Wins per Driver', 
             labels= {'Driver': 'Drivers', 'Race_Wins': 'Number of Wins'},
             color='Driver')
             
Driver_Wins_fig.show()
```

This figure shows that Lewis Hamilton has the higest number of wins with 103, with Michael Schumacher in second place with 91 wins

### Figure comparing race entries and number of wins per driver:

```python
Entries_Wins_fig = px.scatter(F1Drivers_df, x='Race_Entries', y='Race_Wins',color='Driver', title='Number of Race Entries vs Number of Wins',
                              labels={'Race_Entries':'Race Entries', 'Race_Wins':'Race Wins'})
Entries_Wins_fig.show()
```

In this figure there is a very slight positive correlation between the number of entries and number of wins, telling us that the more race entries you have, the more likley you are to have more wins. However there is 2 massive outliers in the scatter, being Lewis Hamilton (Pink) and Michael Schumacher (Dark Blue), with them both having more than 90 wins in less than 350 race entries, compared to Fernando Alonso whi has 359 entries and only 32 wins.

### Creating a new field heading:
Here I have created a new field that creates the average number of wins based on the number of race entries.

```python
F1Drivers_df['Average_Wins'] = F1Drivers_df['Race_Wins'] / F1Drivers_df['Race_Entries']

F1Drivers_df
```

### New Figure using new data field:

Firstly I will order the dataset based on average wins, so that the driver with the highest average will be on the left, going down in ascending order.

```python
F1Drivers_df_sorted = F1Drivers_df.sort_values(by='Average_Wins', ascending=False)

F1Drivers_df_sorted
```

```python
Driver_raceEntries_fig = px.bar(F1Drivers_df_sorted, x='Driver', y='Average_Wins', 
             title='Drivers Average Wins, Based on Number of Race Entries', 
             labels= {'Driver': 'Drivers', 'Average_Wins': 'Average Wins'},
             color='Driver')
             
Driver_raceEntries_fig.show()
```

Here we can see that Juan Manuel Fangio, has the best average number of wins with 0.461 to 3 dp. Interestingly Lewis Hamilton does not have the highest average wins (0.331), even though he has the most amount of wins.

### Figure Showing average wins per Driver Nationality:

```python
Nationality_AverageWins_fig = px.histogram(F1Drivers_df, x='Nationality', y='Average_Wins', 
             title='Number of Wins per Driver', 
             labels= {'Nationality': 'Nationality', 'Average_Wins': 'Average Number of Wins'},
             color='Nationality')
             
Nationality_AverageWins_fig.show()
```

Here we can see that the United Kingdom has the highest average number of wins (2.309) based on their total number of entries. The United States is in second place with an average of 1.855 and Italy is in 3rd place with an average of 1.039.

### Below is a Chlopleth Map highlighting each country that has data for F1 drivers:

First I need to create a new dataframe only having the nationality and the sum of the average wins, so that plotly can correctly read the data

```python
Map_data = F1Drivers_df.groupby('Nationality')['Average_Wins'].sum().reset_index(name='Average_Wins')

Map_data
```

```python
DriverAverage_map_fig = px.choropleth(Map_data, 
                    locations='Nationality', 
                    locationmode='country names',
                    color='Average_Wins', 
                    hover_name='Nationality',
                    color_continuous_scale=px.colors.sequential.Plasma_r,
                    title='Average Wins per Driver by Nationality')

DriverAverage_map_fig.show()
```

This map allows for us to see which parts of the world there is no data and it also highlights countries that there is data for. Hovering over the country on the map will show the user the sum of the average wins of a country, the higher the number of average wins the darker the country will be coloured. The Map clearly shows the F1 is dominated by the west.
